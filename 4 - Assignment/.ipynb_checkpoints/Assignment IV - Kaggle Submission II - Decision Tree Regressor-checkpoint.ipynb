```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
url = '../../Datasets/house-prices-advanced-regression-techniques/train.csv'
train = pd.read_csv(url, sep=",")
train.head(10)
```

```python
y_train = train.SalePrice
y_train.head()
```

```python
corr_abs = train.corr().abs().sort_values(by=['SalePrice'], ascending=False)
mostly_corr = corr_abs[corr_abs.iloc[:,-1] > 0.5]
mostly_corr.columns
#checking any null values
train[mostly_corr.columns].isnull().any()
#mostly_corr.isnull().any()

# filling null values
train[mostly_corr.columns] = train[mostly_corr.columns].fillna(0)
train[mostly_corr.columns].isnull().any()
train[mostly_corr.columns]
```

```python
X_train = train[mostly_corr.columns].drop(['SalePrice', 'Id'], axis=1).copy()
X_train
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
print(dt_model.predict(X_train.head()))
```

```python
from sklearn.metrics import mean_absolute_error
train_predictions = dt_model.predict(X_train)
print(train_predictions)
mean_absolute_error(y_train, train_predictions)
```

```python
test_url = '../../Datasets/house-prices-advanced-regression-techniques/test.csv'
test = pd.read_csv(test_url)
print(test.shape)
test_cols = test[X_train.columns]
test_predictions = dt_model.predict(test_cols.fillna(0))
print(test_predictions)
```

```python
submission = pd.DataFrame({'Id': test.Id, 'SalePrice': test_predictions})
submission.to_csv('decision_tree_submission.csv', index=False)
```
