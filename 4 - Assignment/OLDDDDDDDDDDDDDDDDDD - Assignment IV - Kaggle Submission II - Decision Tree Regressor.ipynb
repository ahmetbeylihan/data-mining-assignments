```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
url = '../../Datasets/house-prices-advanced-regression-techniques/train.csv'
df = pd.read_csv(url, sep=",")
df.head(10)
```

```python
y_train = df.SalePrice
print(y_train.head())
```

```python
belirleyici = ['YearBuilt','1stFlrSF','2ndFlrSF','LotArea','YrSold','FullBath','MSSubClass','TotRmsAbvGrd','BedroomAbvGr']
```

```python
veri_belirleyici = df[belirleyici]
veri_belirleyici.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(veri_belirleyici, y_train)
print(dt_model.predict(veri_belirleyici.head()))
```

```python
from sklearn.metrics import mean_absolute_error
train_predictions = dt_model.predict(veri_belirleyici)
print(train_predictions)
mean_absolute_error(y_train, train_predictions)
```

```python
test_url = '../../Datasets/house-prices-advanced-regression-techniques/test.csv'
test = pd.read_csv(test_url)
test_belirleyici = test[belirleyici]
test_predictions = dt_model.predict(test_belirleyici)
print(test_predictions)
```

```python
submission = pd.DataFrame({'Id': test.Id, 'SalePrice': test_predictions})
submission.to_csv('decision_tree_regressor_submission.csv', index=False)
```
