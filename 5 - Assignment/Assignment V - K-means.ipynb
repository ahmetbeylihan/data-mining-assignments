```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
url = '../../Datasets/iris/iris.data'
data = pd.read_csv(url, sep=",", header=None)
data.head(10)
```

```python
#data.iloc[:,0:-1].values
```

```python
#data.iloc[:,1:-1]
```

```python
#Selecting attributes and scaling..
X = data.iloc[:,0:-1].values
X = StandardScaler().fit_transform(X)
```

**Applying KMeans Algorithm for 5 times**

```python
#Starting iteration from 1 index
iteration_count = 6
number_of_clusters = 3
cluster_of_centers=[]
labels=[]

for i in range(1, iteration_count):
    kmeans = KMeans(n_clusters=number_of_clusters, random_state=1, n_init=1, max_iter=1, init="random").fit(X)
    predict = kmeans.predict(X)
    cluster_of_centers.append(kmeans.cluster_centers_)
    labels.append(kmeans.labels_)
    
print("Cluster Centers\n")
print(cluster_of_centers,'\n')
print("-----------------------------------------------------------------------------------\n")
print("Cluster Labels of each sample\n")
print(labels)
```

**Plotting 2D Figure**

```python
plt.scatter(X[predict==0,0], X[predict==0,1], s=50, color='red', label = 'Iris-setosa')
plt.scatter(X[predict==1,0], X[predict==1,1], s=50, color='blue', label = 'Iris-versicolour')
plt.scatter(X[predict==2,0], X[predict==2,1], s=50, color='green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'gold', label = 'Centroids')
plt.legend(loc='upper right')
plt.title('Iris Dataset K-Means Clustering 2D Visualization')
plt.show()
```

<p>Three clusters are seen clearly</p>

**Plotting 3D Figure**

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:50,0],X[:50,1],X[:50,2],color='red')
ax.scatter(X[50:100,0],X[50:100,1],X[50:100,2],color='blue')
ax.scatter(X[100:150,0],X[100:150,1],X[100:150,2],color='green')
plt.title('Iris Dataset K-Means Clustering 3D Visualization')
plt.show()
```
